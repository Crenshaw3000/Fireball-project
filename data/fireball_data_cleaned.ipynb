```python
#Read file and set the column index 
df = pd.read_csv('cneos_fireball_data.csv')
df
```

```python
#Take the log of Caculated Total Impact Energry
df['energy']= np.log(df['Calculated Total Impact Energy (kt)'])
df
```

```python
#rename columns
df.rename(columns = {'Peak Brightness Date/Time (UT)':'date', 'Latitude (deg.)':'latitude', 'Longitude (deg.)':'longitude'}, inplace = True)
df
```

```python
#drop empty values in place
df.dropna(inplace=True)
```
